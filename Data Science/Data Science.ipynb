```python
# 📦 Import necessary libraries
import pandas as pd
import numpy as np



# 👀 Preview the original columns before normalization
print("Original Price, RAM, and Battery Values:")
display(df[['price', 'ram_capacity', 'battery_capacity']].head())

# 🧮 Apply Min-Max Normalization
df['price_minmax'] = (df['price'] - df['price'].min()) / (df['price'].max() - df['price'].min())
df['ram_minmax'] = (df['ram_capacity'] - df['ram_capacity'].min()) / (df['ram_capacity'].max() - df['ram_capacity'].min())
df['battery_minmax'] = (df['battery_capacity'] - df['battery_capacity'].min()) / (df['battery_capacity'].max() - df['battery_capacity'].min())

# ✅ Display the result
print("📊 After Min-Max Normalization:")
display(df[['price', 'price_minmax', 'ram_capacity', 'ram_minmax', 'battery_capacity', 'battery_minmax']].head(10))
```

```python
print("Original values before normalization:")
print(df[['price', 'ram_capacity', 'battery_capacity']].head())

# 🧮 Step 4: Apply Z-Score Normalization
df['price_zscore'] = (df['price'] - df['price'].mean()) / df['price'].std()
df['ram_zscore'] = (df['ram_capacity'] - df['ram_capacity'].mean()) / df['ram_capacity'].std()
df['battery_zscore'] = (df['battery_capacity'] - df['battery_capacity'].mean()) / df['battery_capacity'].std()

# ✅ Step 5: Show the result
print("\n📊 Z-Score Normalized Data:")
print(df[['price', 'price_zscore', 'ram_capacity', 'ram_zscore', 'battery_capacity', 'battery_zscore']].head(10))
```

```python
# 🧮 Function to apply Decimal Scaling Normalization
def decimal_scaling(column):
    # Drop NaNs temporarily if they exist
    max_abs = column.dropna().abs().max()
    j = len(str(int(max_abs))) if max_abs != 0 else 1  # avoid divide-by-zero
    return column / (10 ** j)

# ✅ Apply Decimal Scaling on numeric columns
df['price_decimal'] = decimal_scaling(df['price'])
df['ram_decimal'] = decimal_scaling(df['ram_capacity'])
df['battery_decimal'] = decimal_scaling(df['battery_capacity'])

# 📊 Display result
print("📊 Decimal Scaling Result:")
print(df[['price', 'price_decimal', 'ram_capacity', 'ram_decimal', 'battery_capacity', 'battery_decimal']].head(10))
```

```python
# 🧪 Step 1: Show original values before binning
print("Original values before binning:")
print(df[['price', 'ram_capacity', 'battery_capacity']].head(10))

# 🧮 Step 2: Apply Equi-Width Binning (4 bins)
df['price_bin_width'] = pd.cut(df['price'], bins=4, labels=["Low", "Medium", "High", "Very High"])
df['ram_bin_width'] = pd.cut(df['ram_capacity'], bins=4, labels=["Low", "Medium", "High", "Very High"])
df['battery_bin_width'] = pd.cut(df['battery_capacity'], bins=4, labels=["Low", "Medium", "High", "Very High"])

# 🧮 Step 3: Apply Equi-Depth Binning safely
# Step 3a: Create depth bin edges with duplicates dropped
battery_bins = pd.qcut(df['battery_capacity'], q=4, duplicates='drop', retbins=True)[1]
# Step 3b: Make matching number of labels
labels = ["Bin " + str(i+1) for i in range(len(battery_bins) - 1)]
# Step 3c: Apply qcut with corrected labels
df['battery_bin_depth'] = pd.qcut(df['battery_capacity'], q=4, labels=labels, duplicates='drop')

# Step 3d: Apply price & RAM depth bins (they work fine with 4 labels)
df['price_bin_depth'] = pd.qcut(df['price'], q=4, labels=["Low", "Medium", "High", "Very High"])
df['ram_bin_depth'] = pd.qcut(df['ram_capacity'], q=4, labels=["Low", "Medium", "High", "Very High"])

# ✅ Step 4: Show result
print("\n📊 Binned Data (Equi-Width and Equi-Depth):")
print(df[[
    'price', 'price_bin_width', 'price_bin_depth',
    'ram_capacity', 'ram_bin_width', 'ram_bin_depth',
    'battery_capacity', 'battery_bin_width', 'battery_bin_depth'
]].head(10))
```

```python
# 🧮 Spearman correlation with 'price'
spearman_corr_price = df.corr(method='spearman', numeric_only=True)['price']

# 📋 Sort by absolute value
spearman_corr_price_sorted = spearman_corr_price.abs().sort_values(ascending=False)

# 🖥️ Display clean table
print("📊 Spearman Correlation with Price (sorted):\n")
print(spearman_corr_price_sorted)
```

```python
edges = df[['brand_name', 'processor_brand']].dropna()

# ✅ Create a directed graph
G = nx.DiGraph()

# ➕ Add edges: brand → processor used
for i, row in edges.iterrows():
    G.add_edge(row['brand_name'], row['processor_brand'])

# 🧠 Compute PageRank
pagerank_scores = nx.pagerank(G)

# 📋 Convert to DataFrame
pagerank_df = pd.DataFrame(pagerank_scores.items(), columns=['Node', 'PageRank']).sort_values(by='PageRank', ascending=False)

# 🖥️ Show top results
print("📊 PageRank (Top nodes):\n")
print(pagerank_df.head(10))
```

```python
edges_camera = df[['brand_name', 'primary_camera_rear']].dropna()

# ✅ Create a directed graph: brand → camera size
G_camera = nx.DiGraph()
for _, row in edges_camera.iterrows():
    brand = row['brand_name']
    camera = str(row['primary_camera_rear'])  # Convert to string in case of numbers
    G_camera.add_edge(brand, camera)

# 🧠 Apply PageRank
pagerank_camera = nx.pagerank(G_camera)

# 📊 Convert to DataFrame for display
pagerank_df = pd.DataFrame(pagerank_camera.items(), columns=['Node', 'PageRank'])

# 🔍 Filter to show only brands (not camera sizes)
brand_ranks = pagerank_df[pagerank_df['Node'].isin(df['brand_name'].unique())]

# ✅ Sort and display top brands by PageRank (camera importance)
brand_ranks = brand_ranks.sort_values(by='PageRank', ascending=False)

print("📸 Smartphone Brands Ranked by Camera PageRank:")
print(brand_ranks.head(10))
```
